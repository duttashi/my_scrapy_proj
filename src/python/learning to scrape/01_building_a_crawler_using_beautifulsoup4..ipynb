Objective: Create a basic crawler that can extract all the links in a page and then browse to each of the extracted links. 

    Status: Work in progress

    Script create date: 25/10/2018

```python
# import the require libraries
from bs4 import BeautifulSoup
import requests
```

```python
url = input("Enter a website to extract the URL's from: ")

r  = requests.get("http://" +url)
data = r.text
# create the soup
soup = BeautifulSoup(data)

for link in soup.find_all('a'):
    print(link.get('href'))
```
